```python
%%writefile reducerfile.py

import sys

currentWord = None
val = 0

for line in sys.stdin:
    try:
        word, value = line.strip().split('\t',1)
        if(word == currentWord):
            val+=1
        else:
            if currentWord:
                print "%s\t%d" % (word,val)
            currentWord = word
            val = 1
    except:
        continue

print "%s\t%d" % (currentWord,val)
```
